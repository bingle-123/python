## 级联
1. np.concatenate()
2. 参数必须是列表  一定要加中括号或小括号
3. 维度必须相同
4. 形状必须相同
5. 级联的方向默认是 shape 这个 tuple 的第一个值所代表的维度方向
6. 可通过 axis 参数改变级联的方向

```python
n1 = np.random.randint(0,10, size=(5,5))
```

```python
n1.shape
```

```python
np.concatenate((n1,n1))
```

```python
np.concatenate((n1,n1), axis=1)
```

### 水平级联和垂直级联

> np.hstack 和 np.vstack

处理自己, 进行维度的变换, 降维(hstack)和升维(vstack)

```python
n2 = np.random.randint(1, 100, 10)
n2
```

```python
# 垂直
# 水平转换为竖直  增加维度
n3 = np.vstack(n2)
```

```python
n3
```

```python
n4 = np.array([[2,3,4,5, 6],[1,2,3,4,5]])
np.hstack(n4) # 降低维度
```

```python
cat.shape
```

```python
# 降为 二维
cat2 = np.hstack(cat)
```

```python
#  二次降维 变为一维
np.hstack(cat2)
```

```python
cat.shape
```

```python
# 在 下标为 1 和 3 前进行切分, 得到的是三个数组,在一个列表中
# 切分默认的是按照最外面的行进行切分
# 默认 axis = 0
# axis = 1  进行切分的是列
np.split(n5, (1,3))
```

## 聚合操作
>  求和  np.sum

> 最大最小  np.max()   np.min()

> 平局值,  np.mean()

> 方差  std

> 最大最小值的索引   np.argmax()  np.argmin()

> np.argwhere(n>50)  查找 数组 n  中 大于50 的索引  得到的是索引 index
通过  n[index]可以取值

```python
np.max(n, axis=0)
```

```python
n.mean()
```

```python
display(cat.max(), cat.min())
```

```python
# axis = 0 将三维数组内的 包含的二维数组看成是 一维的进行取值, 
# 每行对应位置上的进行比较取最大值, 最后结构还是要表示成二维数组
n.max(axis=0)

# axis = 1 对内部每个二维数组的每一行对应的位置上的进行比较取值
n.max(axis=1)

# axis = 2  对最里面的一维数组进行取值
```
